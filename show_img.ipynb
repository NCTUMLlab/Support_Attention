```python
import numpy as np
from scipy import ndimage
import skimage.transform
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
alps = np.load('data/visual/alps.npy')
sam_caption = np.load('data/visual/sample_captions.npy')
image_file = np.load('data/visual/image_name.npy')
```

```python
idx=61
img = ndimage.imread(image_file[idx])
plt.imshow(img)
print sam_caption[idx]
```

```python
idx=70
print sam_caption[idx]
img = ndimage.imread(image_file[idx])
plt.figure(figsize=(30,30))
plt.subplot(4,4,1); plt.imshow(img)
plt.axis('off')

words = sam_caption[idx].split(" ")
for t in xrange(len(words)):
    if t > 18:
        break
    plt.subplot(4, 4, t+2)
    plt.text(0, 1, '%s'%(words[t]) , color='black', backgroundcolor='white', fontsize=50)
    plt.imshow(img)
    alp_curr = alps[idx,t,:].reshape(14,14)
    alp_img = skimage.transform.pyramid_expand(alp_curr, upscale=16, sigma=20)
    plt.imshow(alp_img, alpha=0.85, cmap='gray')
    plt.axis('off')
plt.show()
```
